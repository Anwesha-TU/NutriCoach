```python
import json
from sentence_transformers import SentenceTransformer
```

```python
def ing_to_text(ingredient):
    concerns = ", ".join(ingredient["health_concern_type"]).lower()

    text = f"""{ingredient["name"]}.
Category: {ingredient["category"]}.
Signals {ingredient["processing_signal"].lower()} processing.
Evidence is {ingredient["evidence_strength"].lower()}, with concerns related to {concerns}.
{ingredient["context_notes"]}.
Primarily used to {ingredient["typical_role"].lower()}.
Relevant because it {ingredient["consumer_relevance"].lower()}.
"""

    return text
```

```python
with open("ingredients.json", "r") as f:
    ingredients = json.load(f)

for i, ing in enumerate(ingredients):
    ingredients[i]["text"] = ing_to_text(ing)
```

```python
print(ingredients[0]["text"])
```

```python
# embedding model - https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
for ing in ingredients:
    ing["embedding"] = model.encode(ing["text"]).tolist()
```

```python
with open("embeddings.json", "w") as f:
    json.dump(ingredients, f, indent=4)
```
